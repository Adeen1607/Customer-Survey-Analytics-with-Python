```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your Excel files
department_data = pd.read_excel('department_results_expanded.xlsx')
product_data = pd.read_excel('product_results_expanded.xlsx')
```

```python
# Sum of visits for each department and purchases for each product
department_totals = department_data.iloc[:, 2:].sum()
product_totals = product_data.iloc[:, 2:].sum()
```

```python
# Histogram of department visit counts
plt.figure(figsize=(12, 7))
department_totals.plot(kind='bar', color='skyblue')
plt.xlabel('Departments')
plt.ylabel('Visit Counts')
plt.title('Distribution of Department Visit Counts')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Histogram of product purchase counts
plt.figure(figsize=(12, 7))
product_totals.plot(kind='bar', color='lightgreen')
plt.xlabel('Products')
plt.ylabel('Purchase Counts')
plt.title('Distribution of Product Purchase Counts')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Top 5 most visited departments
top_5_most_visited_depts = department_totals.nlargest(5)
plt.figure(figsize=(10, 6))
top_5_most_visited_depts.plot(kind='bar', color='dodgerblue')
plt.xlabel('Departments')
plt.ylabel('Visit Counts')
plt.title('Top 5 Most Visited Departments')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Top 5 most purchased products
top_5_most_purchased_products = product_totals.nlargest(5)
plt.figure(figsize=(10, 6))
top_5_most_purchased_products.plot(kind='bar', color='seagreen')
plt.xlabel('Products')
plt.ylabel('Purchase Counts')
plt.title('Top 5 Most Purchased Products')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Pie chart for department visits
plt.figure(figsize=(8, 8))
department_totals.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.ylabel('')
plt.title('Share of Department Visits')
plt.show()
```

```python
# Pie chart for product purchases
plt.figure(figsize=(8, 8))
product_totals.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.ylabel('')
plt.title('Share of Product Purchases')
plt.show()
```
